```python
print(img.shape) # shape of the image ie (1280, 858, 3) => (height, width, channels)
# Height => Rows in pixels (upar se neeche ka size)
# Width => 	Columns in pixels (left to right ka size)
# Color info: 3 (BGR/RGB), 1 (grayscale), 4 (RGBA with alpha)
# Data type of pixel values => unit8
# Pixel values => Each channel’s intensity (0 = black, 255 = max brightness)
# RGB =>  Red, Green, Blue
# Pure Red => [255,0,0]
# Pure Green => [0,255,0]
# Pure Blue => [0,0,255]
# RGB Format Pure Red = [255,0,0]
# Red channel → 255 (full bright red)
# Green channel → 0 (green light band)
# Blue channel → 0 (blue light band) 
# But BGR => Blue, Green, Red
# In BGR [255,0,0] is pure Blue
# BGR Format Pure Green = [0,255,0]
# BGR Format Pure Red = [0,0,255]
# Matplotlib RGB expect karta hai aur OpenCV BGR 
```

```python
# Step 1: Read image, show in notebook, and print basic numeric info
import cv2
import matplotlib.pyplot as plt

# replace with your file name or full path if needed
img_path = "img1.jpg"

# read (OpenCV reads as BGR)
img_bgr = cv2.imread(img_path)

# safety check
if img_bgr is None:
    raise FileNotFoundError(f"Image not found at: {img_path}  -- check path!")

# basic numeric info
print("type:", type(img_bgr))
print("shape (H, W, C):", img_bgr.shape)   # e.g. (480, 640, 3)
print("dtype:", img_bgr.dtype)             # usually uint8
print("min/max pixel value:", img_bgr.min(),"/", img_bgr.max())

# convert to RGB for matplotlib display (so colors are correct)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_grey = cv2.cvtColor(img_bgr, cv2.COLOR_RGB2GRAY)
# display
plt.figure(figsize=(8,6))
plt.imshow(img_bgr)
plt.axis('off')
plt.title(f"{img_path}   shape: {img_bgr.shape}")
plt.show()
```

```python
img_RGBA = cv2.cvtColor(img_grey, cv2.COLOR_GRAY2RGBA)
# Grayscale image ka format
# Shape: (H, W)  3rd dimension (channels) hota hi nahi
# Har pixel: ek single number (0255 range)
# 0  pure black
# 255  pure white
# beech ka koi bhi value  grey ka koi shade
```

```python
plt.imshow(img_RGBA)
plt.axis('off')
```

```python
import numpy as np
import matplotlib.pyplot as plt

pixel = np.array([[[0, 0, 100]]], dtype=np.uint8)  # shape: (1, 1, 3)
# [[[0, 0, 100]]]  3D array (height=1, width=1, channels=3)
plt.imshow(pixel)
plt.axis('off')
plt.show()
```

```python
# RGBA format pixel example
# RGBA ek color format hai jo Red, Green, Blue, Alpha channels ka combination hota hai.
# Alpha channel opacity (transparency) ko represent karta hai.
# 0 ka matlab fully transparent (bilkul dikhayi nahi deta).
# 255 ka matlab fully opaque (pura dikhayi deta).
pixel = np.array([0, 0, 100, 0])  # RGBA format
plt.imshow(pixel.reshape(1,1,4)) # reshape to 1x1 pixel
plt.axis('off')
# toh white dikhne ka reason ye hai ke alpha channel 0 hai, yani pixel fully transparent hai.
#  Matlab ye pixel dikhta hi nahi, uske peeche ka background dikh raha hai, jo matplotlib me default white hota hai.

```
